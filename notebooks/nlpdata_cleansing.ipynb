```python
import numpy as np
import pandas as pd
import re
import seaborn as sns
import sklearn as sk
import sklearn.model_selection
```

The notebook relies on `dir` to locate the directory storing our data, in the form of csv files

```python
dir = '../data'
```

# load data

```python
df = pd.read_csv( f'{dir}/nlpdata-8606.csv', encoding='utf8' )
### remove "special" characters
for c in df.columns:
  df[ c ] = df[ c ].apply( lambda x: re.sub( r' *[^\x00-\x7F]+', '', x ) if pd.notnull( x ) else x ).astype( 'string' )
print( df.shape )
df.head()
```

```python
df[ df.Question.isna() | df.Answer.isna() ]
```

```python
df.dropna( subset=[ 'Question', 'Answer' ], inplace=True )
print( df.shape )
df.head()
```
