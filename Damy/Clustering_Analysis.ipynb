```python
import ast
import numpy as np
import pandas as pd
from sklearn.cluster import KMeans
from sentence_transformers import SentenceTransformer, util
```

```python
def text_embed(text_col_name, dataframe, vector_col_name):
    model = SentenceTransformer("thenlper/gte-small")
    dataframe[text_col_name] = dataframe[text_col_name].astype(str)
    dataframe[vector_col_name] = dataframe[text_col_name].apply(lambda x: model.encode(x))
    return dataframe
```

# CLIMATE  ACTION

```python
climate_df = pd.read_csv(r"C:\Users\RedHat\Downloads\Tiiqu\Unstriped Climate Change for SUPERCLEANED dataset 23-3-2024.csv")
climate_df = climate_df.iloc[:, 1:]
climate_df
```

```python
climate_df_unallocated = climate_df[climate_df['Topic'].isna()].reset_index(drop=True)
climate_df_unallocated
```

```python
# subtopic embedding for smart cities
text_embed('QA_prep', climate_df_unallocated, 'QA_prep_vector')
```

```python
#climate_df_unallocated['QA_prep_vector']
```

```python
corpus = climate_df_unallocated['QA_prep_vector']
corpus_array = np.vstack(corpus)
```

```python
corpus
```

```python
#print(corpus_array)
```

```python
print(corpus.shape)
```

```python
print(corpus_array.shape)
```

```python
# Fit KMeans model
kmeans = KMeans(n_clusters=10, random_state=42, n_init=20)
climate_df_unallocated['Cluster'] = kmeans.fit_predict(corpus_array)

# Display the resulting DataFrame
climate_df_unallocated
```

```python
climate_df_unallocated = climate_df_unallocated[['Question', 'Answer', 'Macrotopic ', 'Topic',
                                                   'Assigned Subtopic(Preprocessed)', 'Source Url', 'Label ', 'Cluster']]
```

```python
climate_df_unallocated.to_csv('Clusters of Unallocated Climate dataset 01-04-2024.csv')
```
